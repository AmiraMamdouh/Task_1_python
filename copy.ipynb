```python
def cleaning(text):
    my_lst=[]
    flag= False
    if type(text) == float: #this condition is necessary to make sure 
        #that the NaN values are skipped from being used in itteration
        clean_text=text
    else:
        for charac in text:
            if charac.isalpha() == False and charac != ' ' :
                flag= True #found a  character which is not an alpha and not a space, then stop itterating
            if(flag == False):
                my_lst.append(charac)
        clean_text=''.join(my_lst)
    return clean_text
```
